# Resmi Göstermek için Fonksiyon Yazdım

```python
def resmi_gosterme(image_path):

    image = cv2.imread(image_path)
    if image is None:
        print("Error: Image not found")
        return

    cv2.imshow("YOLO Sonucu",image)

    cv2.waitKey(0)
    
    
    cv2.destroyAllWindows()
```

# Yolo Nano Model
